```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
models = ["DALLÂ·E 3", "Stable Diffusion", "Imagen", "MidJourney"]
fid_scores = [12.5, 15.2, 11.7, 18.4]
inception_scores = [8.2, 7.8, 8.5, 7.5]
text_image_alignment = [8.5, 7.0, 9.0, 6.0]  # Mapping High = ~9, Moderate = ~7
inference_speed = [2.1, 0.8, 3.2, 1.7]
```

```python
# First plot: FID Scores vs Inception Scores
fig, ax = plt.subplots(figsize=(8,5))
x = np.arange(len(models))
width = 0.35
ax.bar(x - width/2, fid_scores, width, label='FID Scores', color='royalblue')
ax.bar(x + width/2, inception_scores, width, label='Inception Scores', color='seagreen')

ax.set_xticks(x)
ax.set_xticklabels(models)
ax.set_ylabel("Scores")
ax.legend()
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(8,5))

ax1.set_xlabel("Models")
ax1.set_ylabel("Text-Image Alignment (Higher is Better)", color="blue")
ax1.plot(models, text_image_alignment, marker='o', color="blue")
ax1.tick_params(axis="y", labelcolor="blue")

ax2 = ax1.twinx()
ax2.set_ylabel("Inference Speed (Seconds)", color="red")
ax2.plot(models, inference_speed, marker='o', linestyle='dashed', color="red")
ax2.tick_params(axis="y", labelcolor="red")

plt.title("Text-Image Alignment vs. Inference Speed")
plt.show()
```
